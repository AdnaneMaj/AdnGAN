## Transform GIF to JPG

```python
from PIL import Image
import os

# Specify the input and output directories
input_dir = "./Drive_data/1st_manual"
output_dir = "./Drive_data/1st_manuall"

# Create the output directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Loop through all files in the input directory
for filename in os.listdir(input_dir):
    # Check if the file is a GIF image
    if filename.endswith(".gif"):
        # Open the GIF image
        gif_image = Image.open(os.path.join(input_dir, filename))
        
        # Convert GIF to JPEG
        jpeg_image = gif_image.convert("RGB")
        
        # Generate output filename by replacing ".gif" with ".jpg"
        output_filename = os.path.splitext(filename)[0] + ".jpg"
        
        # Save the JPEG image to the output directory
        jpeg_image.save(os.path.join(output_dir, output_filename))
        
        print(f"Converted {filename} to {output_filename}")

print("Conversion completed.")
```
